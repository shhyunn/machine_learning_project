```python
df_result = pd.concat([df1, df2], ignore_index = True)
df_result
```

```python
X = df_result['x']
y = df_result['y']
```

```python
len(X), len(y)
```

```python
X[1]
```

```python
import nltk
nltk.download('names')
nltk.download('wordnet')

from nltk.corpus import names
all_names = set(names.words())

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
clean_X = []
for i in range(len(X)):
  text = X[i]
  clean_text = []
  for word in text.split(' '):
    word = word.lower()
    if word.isalpha() and word not in all_names and len(word) > 2:
      clean_text.append(lemmatizer.lemmatize(word))
  clean_X.append(' '.join(clean_text))
```

```python
len(clean_X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(clean_X, y, test_size = 0.3, random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english', max_features = 500) #불용어 처리
term_docs_train = cv.fit_transform(X_train)
term_docs_test = cv.transform(X_test)
term_docs_train = term_docs_train.toarray()
```

```python
term_docs_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB(alpha = 1.0, fit_prior=True)
clf.fit(term_docs_train, Y_train)
```

```python
accuracy = clf.score(term_docs_test, Y_test)
print('The accuracy using MultinomialNB is: {0:.1f}%'.format(accuracy*100))
```

```python
y_pred = clf.predict(term_docs_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, Y_test))
```
